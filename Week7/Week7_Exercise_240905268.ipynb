```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question 1 
# For the IPL dataset, answer the specified questions with summarization and effective visuals using Matplotlib 
# & Seaborn libraries

var_dfIPL = pd.read_csv("matches.csv")

var_dfIPL
```

```python
var_dfIPL.drop(['umpire3'],axis = 1, inplace = True)
```

```python
var_dfIPL
```

```python
# 1. Find the names of teams that are part of IPL.

var_arrIplTeams = var_dfIPL['team1'].unique()

for team in var_arrIplTeams:
    print(team)
```

```python
# 2. Visualize the total number of matches won by each team.
var_Wins = var_dfIPL[var_dfIPL['result'] == 'normal']

var_TeamWins = var_Wins['winner'].value_counts()

var_TeamWins.plot(kind = 'bar',color = 'blue', figsize =(10,10))
```

```python
# 3. List the names of umpires who have served as first umpire and the number of matches.

var_FirstUmpires = var_dfIPL['umpire1'].value_counts()

for name in var_FirstUmpires.index:
    print(name)
    
var_FirstUmpires.plot(kind = 'bar', color = 'red', figsize = (10,10))
```

```python
# 4. Visualize how many matches were held in each venue.

var_MatchesHeld = var_dfIPL['venue'].value_counts()

var_MatchesHeld.plot(kind = 'bar',color = 'green', figsize = (10,10))
```

```python
# 5. Visualize the top 10 players identified as the player of the match for the most significant number of matches.

var_PlayerOfMatch = var_dfIPL['player_of_match'].value_counts()

var_Top10 = var_PlayerOfMatch.iloc[:10]

var_Top10.plot(kind = 'bar',color = 'yellow', figsize = (10,10))
plt.title("Top 10 Player of Match")
plt.xlabel("Players")
plt.ylabel("Number of Player of match")
plt.show()
```

```python
print(var_MatchesHeld)
```

```python
# 6. Visualize the bottom 10 venues where the least number of matches were played.a
var_Bottom_10_Venue = var_MatchesHeld[-10:]

var_Bottom_10_Venue.plot(kind = 'bar', color = 'purple', figsize = (10,10))
```

```python
# 7. Visualize the number of wins of each team for the venue of ‘MA Chidambaram Stadium,Chepauk’

var_Venue_MA_Stadium = var_dfIPL[var_dfIPL['venue'] == 'MA Chidambaram Stadium, Chepauk']

var_Venue_MA_Stadium_Wins = var_Venue_MA_Stadium['winner'].value_counts()

var_Venue_MA_Stadium_Wins.plot(kind = 'bar', color = 'lime', figsize = (10,10))
```

```python
# 8. Find and list the venues where the match resulted in a tie.

var_dfTies = var_dfIPL[var_dfIPL['result'] == 'tie']

for venue in var_dfTies['venue'].unique():
    print(venue)
```

```python
# 9. List and visualize the best teams(top 5), season-wise

var_dfTrueWins =var_dfIPL[var_dfIPL['result'] == 'normal']

var_dfSeasonWise = var_dfTrueWins.groupby('season')['winner'].value_counts().groupby('season').head(5)

var_dfSeasonWise
```

```python
# 10. Analyse and visualize the relationship between winning the toss and winning the match.

var_dfIPL['Toss&Win'] = var_dfIPL['toss_winner'] == var_dfIPL['winner']

var_toss_and_win = var_dfIPL['Toss&Win'].value_counts(normalize = True) * 100
plt.figure(figsize = (10,10))
plt.pie(var_toss_and_win,labels = ['Lost Match ,Won Toss','Won Match,Won Toss'],autopct = '%1.1f%%', colors =['red','blue'], startangle = 0, explode = (0.1,0))
plt.show()
```
