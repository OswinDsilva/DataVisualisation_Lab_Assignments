```python
# Q2
# 2. For the CEREALS dataset, perform data preprocessing and answer the following questions.
# a. Create a table with the 5 number summary of all the numeric attributes.
# b. For each of the numeric attributes (proteins upto vitamins) , identify and replace all missing data(indicated with -1) with the arithmetic mean of the attribute.
# c. Create a table with the 5 number summary of all the numeric attributes after treating missing values. Do you think the strategy used in dealing with missing values was effective?
# d. For each numeric attribute (proteins upto vitamins), identify and replace all noisy data with the median of attribute.
# e. Create a table with the 5 number summary of all the numeric attributes after treating noisy values. 
# Do you think the strategy used in dealing with noisy values was effective?
```

```python
# Reading the file into a dataframe

var_dfCereals = pd.read_csv("Cereals.csv")

# Printing the dataframe

var_dfCereals
```

```python
# 5 number summary of all the numeric attributes.  

var_dfCereals.describe()
```

```python
# replacing all missing data(indicated with -1) with the arithmetic mean of the attribute

# Replacing all the -1 in the dataframe by NaN

var_dfCereals.replace(-1,np.nan, inplace = True)

# Replacing the NaN by the mean for each column

var_dfCereals.iloc[:,4:12] = var_dfCereals.iloc[:,4:12].fillna(var_dfCereals.iloc[:,4:12].mean(),axis = 0)

# Printing the new dataframe 

var_dfCereals
```

```python
# 5 number summary of all the numeric attributes after treating missing values.  

var_dfCereals.describe()
```

```python
# Cleaning the outliers

# Using the bound conditions (25% quartile - 1.5*IQR, 75% quartile + 1.5*IQR)

# Looping through columns 4 through 11, which represents the columns - proteins to vitamins
for i in range(4,12):

    # Converting the values to float data type because I was getting a type incompatibility warning
    
    var_dfCereals.iloc[:,i] = var_dfCereals.iloc[:,i].astype('float64')

    #Calculating the Quartiles and the IQR

    var_25Quartile = var_dfCereals.iloc[:,i].quantile(0.25)
    var_75Quartile = var_dfCereals.iloc[:,i].quantile(0.75)
    var_IQR = var_75Quartile - var_25Quartile

    # Calculating the bounds

    var_LowerLimit = var_25Quartile - 1.5*var_IQR
    var_UpperLimit = var_75Quartile + 1.5*var_IQR

    # Replacing outliers

    var_dfCereals.loc[((var_dfCereals.iloc[:,i] < var_LowerLimit) | (var_dfCereals.iloc[:,i] > var_UpperLimit)),var_dfCereals.columns[i]] = var_dfCereals.iloc[:,i].median()

```

# There has been a change in the 5 number summaries of some of the columns especially their means. This change implies that some of the values that were much larger or much smaller compared to the values following the general trend of the data were replaced . In our case , we replaced it by the median bringing the values closer together , thus resulting in a change in mean. These larger/smaller values were the noisy values or also known as outliers.
