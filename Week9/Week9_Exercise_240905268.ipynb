```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For the Sentiment data set, answer the questions provided with effective visualisations wherever asked

var_dfData = pd.read_csv("sentiment.csv", index_col = 0)

var_dfData
```

```python
# Checking Feature types and checking for Null values

var_dfData.info()
```

```python
# 1.In the sentiment feature and text, trim white spaces and convert to lowercase.

# Using .strip() to remove extra whitespaces and using .lower() to convert the text into lower case
var_dfData['Sentiment'] = var_dfData['Sentiment'].str.strip().str.lower()

# Using .strip() to remove extra whitespaces and using .lower() to convert the text into lower case
var_dfData['Text'] = var_dfData['Text'].str.strip().str.lower()

var_dfData
```

```python
# 2. What are the sentiment categories? Tabulate and visualize the number of texts for each category

# Converting the Sentiment feature into categorical data type 
var_dfData['Sentiment'] = var_dfData['Sentiment'].astype("category")

# Obtaining the Categories using .categories
var_SentCategory = var_dfData['Sentiment'].cat.categories

var_SentCategory
```

```python
# Using value_counts() to obtain a count of each category in Sentiment feature
var_Sentiment_Count = pd.DataFrame(var_dfData['Sentiment'].value_counts())

var_Sentiment_Count
```

```python
# Plotting a bar graph of Frequency of the each Sentiment category
# Due to a large number of categories, the labels appear overlapped/squished together.
plt.figure(figsize=(10,10))
sns.barplot(x=var_Sentiment_Count.index,y='count',data=var_Sentiment_Count)
plt.title("Number of Sentiments of each category")
plt.xlabel("Sentiments")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Tabulate and Visualize the text length for each sentiment category

# Obtaining length of each text using .len()
var_dfData['Len'] = var_dfData['Text'].str.len()

# Obtaining average text length for each Sentiment category
var_TextLen = var_dfData.groupby('Sentiment',observed=False)[['Len']].mean()

var_TextLen
```

```python
# Plotting Average length of text of each category
# Due to a large number of categories, the labels appear overlapped/squished together.
plt.figure(figsize=(10,10))
sns.barplot(x='Sentiment',y='Len',data=var_TextLen)
plt.title("Average text length for each category")
plt.xlabel("Sentiments")
plt.ylabel("Average Length")
plt.show()
```

```python
# 4. Find and display the text(s) with the maximum number of likes.

# Finding the maximum value using .max()
var_MaxLikes = var_dfData['Likes'].max()

# Obtaining a dataframe of data which have likes equal to the maximum value calculated prior
var_HighestLikes = var_dfData[var_dfData['Likes'] == var_MaxLikes][['Text','Likes']]

var_HighestLikes
```

```python
# 5. Find and display the text(s) with the maximum reposts. what is its sentiment?

# Obtaining max value using .max()
var_MaxReposts = var_dfData['Retweets'].max()

# Obtaining a dataframe of data which have Retweets equal to the maximum value calculated prior
var_HighestReposts = var_dfData[var_dfData['Retweets'] == var_MaxReposts][['Text','Sentiment','Retweets']]

var_HighestReposts
```

```python
# 6. Tabulate and Visualize the top 5 countries that have posted the most texts with a positive sentiment.

# Removing whitespaces for preventing repetition of countries due to irregular spacing 
var_dfData['Country'] = var_dfData['Country'].str.strip()

# Obtaining only those texts which are labelled as having a 'positive' sentiment
var_PositiveSentiments = var_dfData[var_dfData['Sentiment'] == "positive"]

# Obtaining a Count of number of positive texts according to the Country
var_TopPositiveCountries = pd.DataFrame(var_PositiveSentiments['Country'].value_counts().head(5))

var_TopPositiveCountries
```

```python
# Plotting Countries with Most number of Positive texts using a Bar Graph
plt.figure(figsize=(10,10))
sns.barplot(x='Country',y='count',data=var_TopPositiveCountries)
plt.title("Top 5 Countries with highest Positive Texts")
plt.xlabel("Countries")
plt.ylabel("Frequency")
plt.show()
```

```python
# 7. Find and display all the text(s) with the word â€˜Travelâ€™ in their hashtag.

# Obtaining a dataframe containing data which have 'Travel' in their Hashtags using .contains()
var_TravelHash = var_dfData[var_dfData['Hashtags'].str.contains("Travel")]

var_TravelHash
```

```python
# 8. Find and display the most popular (most number of likes) posts (top 5) from the year 2020 which were of a negative sentiment.

# Obtaining a dataframe of text which are labelled as having 'negative' sentiments from the year 2020
var_NegativeSentiments2020 = var_dfData[(var_dfData['Sentiment'] == "negative") & (var_dfData['Year'] == 2020)]

# Sorting in ascending order by Likes
var_Popular = var_NegativeSentiments2020.sort_values('Likes')

var_Popular
```

```python
# 9. Find and display a list of Users and their posts whose texts are labeled â€˜Happinessâ€™ sentiment.

# Obtaining a dataframe of text which are labelled as having the 'Happiness' sentiment
var_Happiness = var_dfData[var_dfData['Sentiment'] == 'happiness'][['Text','User','Sentiment']]

var_Happiness
```

```python
# 10. Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets with sentiments, â€˜Positive,â€™ â€˜Negative,â€™ and â€˜Neutralâ€™ only.

# Obtaining a dataframe containing only texts having sentiments labelled as 'positive','negative' or 'Neutral'. 
var_Sentiments = var_dfData[(var_dfData['Sentiment'] == 'positive') | (var_dfData['Sentiment'] == 'negative') | (var_dfData['Sentiment'] == 'neutral')]

# Tabulating the frequency of these texts according to the year and the country
var_ct = pd.crosstab(var_Sentiments['Country'],var_Sentiments['Year'])

var_ct
```

```python
# Heatmap plot of the above tabulation
plt.figure(figsize=(10,10))
sns.heatmap(var_ct,annot=True,cmap = "Blues")
plt.title("Heatmap of Number of texts according to Country and Year")
plt.xlabel("Years")
plt.ylabel("Countries")
plt.show()
```

```python
# Array containing all the unique years in the data
var_years = var_Sentiments['Year'].unique()

# Plotting a heat map of Frequency of each Sentiment ('Positive','Negative' and 'Neutral') according to country for each year
# A for loop iterating through each year 
for year in var_years:

    # Obtaining a dataframe of data belonging to the particular year
    var_dfYear = var_Sentiments[var_Sentiments['Year'] == year]

    # Tabulating the data
    var_ct2 = pd.crosstab(var_dfYear['Country'],var_dfYear['Sentiment'])

    # Plotting the heatmap
    plt.figure(figsize=(10,10))
    sns.heatmap(var_ct2,annot=True, cmap ="Greens")
    plt.title(f"Heatmap of Country and the Sentiments for the year {year}")
    plt.xlabel("Sentiments")
    plt.ylabel("Countries")
    plt.show()
```
