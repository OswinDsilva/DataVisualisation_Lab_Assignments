```python
# Question 1
# Generate a 3x4 NumPy array with random integers between 1and 50.
# a. Calculate and print the Mean, Median, and Standard Deviation of the array
# b. Print the Sum of all elements and the sum of each row.
# c. Reshape the 3x4 array into a 2x6 array and print it.

# Generation of the array
var_arrA = np.random.randint(1,51,size = (3,4))
print("Original Array:")
print(var_arrA)
print(f"Dimension of the array is : {var_arrA.shape}\n")


# Part a. Mean, Median, and Standard deviation
print(f"Mean = {np.mean(var_arrA)}")
print(f"Median = {np.median(var_arrA)}")
print(f"Standard Deviation = {np.std(var_arrA)}\n")


# Part b. Sum of elements
print(f"Sum of all elements = {np.sum(var_arrA)}")
for i in range(var_arrA.shape[0]):
    print(f"Sum of row {i+1}:{np.sum(var_arrA[i])}")


# Part c. Reshaping the array
var_arrA = var_arrA.reshape(2,6)
print("\nArray after reshaping :")
print(var_arrA)
```

```python
# Question 2
# 2. Create two (3 * 3) matrices using NumPy and print it. 
# Perform and print the results of the following linear algebra operations
# a. Matrix addition
# b. Matrix subtraction
# c. Matrix multiplication (element-wise and dot product)
# d. Transpose of a matrix
# e. Determinant and inverse (if applicable)


# Generation of array 
var_arrB = np.random.randint(50,size = (3,3))
var_arrC = np.random.randint(50,size = (3,3))
print("Array 1:")
print(var_arrB)
print("\nArray 2:")
print(var_arrC)

# Part a. Matrix Addition
print(f"\nSum of the two arrays is \n{var_arrB + var_arrC}")

# Part b. Matrix Subtraction
print(f"\nDifference of the two arrays is \n{var_arrB - var_arrC}")

# Part c. Matrix Multiplication
print(f"\nElementwise multiplication of the two arrays is \n{var_arrB * var_arrC}")
print(f"\nMultiplication of the two arrays is \n{var_arrB @ var_arrC}")


# Part d. Matrix Tranpose
print(f"\nTranpose of the first array is \n{var_arrB.T}")
print(f"\nTranpose of the second array is \n{var_arrC.T}")

# Part e. Determinant and inverse
print(f"\nDeterminant of first array is {np.linalg.det(var_arrB)}")
if(np.linalg.det(var_arrB)!=0):
    print(f"\nInverse of first array is \n{np.linalg.inv(var_arrB)}")
print(f"\nDeterminant of second array is {np.linalg.det(var_arrC)}")
if(np.linalg.det(var_arrC)!=0):
    print(f"\nInverse of second array is \n{np.linalg.inv(var_arrC)}")
```
